Базовый EDA датасета

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib import rcParams

# figure size in inches
rcParams['figure.figsize'] = 14, 10
plt.rcParams["font.family"] = "serif"
```

```python
dir = 'drive/My Drive/Colab Notebooks/CSC/Projects/REZ_8/'
df = pd.read_csv(dir + 'flats.csv', index_col = 0)
```

```python
ax = sns.countplot(x='name', data=df, order = df['name'].value_counts().iloc[:5].index)
ax.set_title('Распределения типа квартир', fontsize=20)
ax.set_xlabel('name', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
ax = sns.countplot(x="area", data=df, order = df['area'].value_counts().iloc[:5].index)
ax.set_title('Распределения области квартир', fontsize=20)
ax.set_xlabel('area', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
plt.xticks(rotation=-80)
ax = sns.countplot(x="district", data=df, order = df['district'].value_counts().iloc[:25].index)
ax.set_title('Распределения района квартир', fontsize=20)
ax.set_xlabel('district', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
plt.xticks(rotation=-80)
ax = sns.countplot(x="unit", data=df, order = df['unit'].value_counts().iloc[:25].index)
ax.set_title('Распределения города/деревни квартир', fontsize=20)
ax.set_xlabel('unit', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
plt.xticks(rotation=-80)
ax = sns.countplot(x="company", data=df, order = df['company'].value_counts().iloc[:25].index)
ax.set_title('Распределения компаний, продавших квартиры', fontsize=20)
ax.set_xlabel('company', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
df['agent?'] = (pd.isnull(df['agent']) == False)
ax = sns.countplot(x="agent?", data=df, order = df['agent?'].value_counts().iloc[:25].index)
ax.set_title('Распределения квартир проданных при посредничестве агента и нет', fontsize=20)
ax.set_xlabel('agent?', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
df.drop('agent?', inplace=True, axis=1)
```

```python
ax = sns.histplot(df, x="total_area", binrange = [0, 250])
ax.set_title('Распределение общей площади квартир', fontsize=20)
ax.set_xlabel('total_area', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
ax = sns.histplot(df, x="living_area", binrange = [0, 150])
ax.set_title('Распределение жилой площади квартир', fontsize=20)
ax.set_xlabel('living_area', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
ax = sns.histplot(df, x="kitchen_area", binrange = [0, 70])
ax.set_title('Распределение площади кухни квартир', fontsize=20)
ax.set_xlabel('kitchen_area', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
ax = sns.histplot(df, x='ceiling_height', binrange = [2.4, 3])
ax.set_title('Распределение высоты потолков квартир', fontsize=20)
ax.set_xlabel('height', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
ax = sns.histplot(df, x="last_price", binrange = [500000, 70000000])
ax.set_title('Распределение цены продажи квартир', fontsize=20)
ax.set_xlabel('price', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```

```python
df['last_seen'] = pd.to_datetime(df['last_seen'], dayfirst = True)
df['first_seen'] = pd.to_datetime(df['first_seen'], dayfirst = True)
df['time_to_sell'] = (df['last_seen'] - df['first_seen']).dt.days
df.loc[df['last_seen'] == '18-05-21', 'time_to_sell'] = 50
# значение 50 - будет означать, что квартира еще не продана
ax = sns.histplot(df, x="time_to_sell")
ax.set_title('Распределение времени продажи квартир', fontsize=20)
ax.set_xlabel('time', fontsize=20)
ax.set_ylabel('count', fontsize=20)
ax.tick_params(labelsize=13)
```
