Примеры наблюдений из json-файла

```python
from google.colab import drive
drive.mount('/content/drive')
import json
import random
```

```python
dir = 'drive/My Drive/Colab Notebooks/CSC/Projects/REZ_8/'
with open(dir + 'full_db(18-05-21).json') as json_file:
    flats = json.load(json_file)
```

```python
for id in random.sample(flats.keys(), 2):
    print(id, ':')
    for feature in flats[id]:
        print(feature, ': ', flats[id][feature])
    print('----------')
```
